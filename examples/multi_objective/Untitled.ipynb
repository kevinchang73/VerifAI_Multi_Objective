```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

infile = open('result.log', 'r') # *.txt
mode = 'multi' # multi / single
order = 'alternate' # alternate / sequential

result_count_0 = np.zeros(shape=(2,4), dtype=int) # result_count_0[i] = [count of 00, 01, 10, 11 in segment 0] sampled from sampler i
result_count_1 = np.zeros(shape=(2,4), dtype=int) # result_count_1[i] = [count of 00, 01, 10, 11 in segment 1] sampled from sampler i
curr_source = 0
lines = infile.readlines()
infile.close()
counter_examples = []
violations = []
for i in range(len(lines)):
    if mode == 'multi':
        if 'Rho' in lines[i]:
            line = lines[i].strip()
            seg1 = line[line.find('[[')+2:line.find(']')].split(' ')
            val1 = []
            val1_num = []
            for s in seg1:
                if s != '':
                    val1.append(float(s) < 0)
                    val1_num.append(float(s))
            #print(val1)
            #print(val1_num)
            assert len(val1) == 4, 'Invalid length of rho'
            result_count_0[curr_source][val1[1]*2 + val1[3]*1] += 1

            line = lines[i+1].strip()
            seg2 = line[line.find('[')+1:line.find(']]')].split(' ')
            val2 = []
            val2_num = []
            for s in seg2:
                if s != '':
                    val2.append(float(s) < 0)
                    val2_num.append(float(s))
            #print(val2)
            #print(val2_num)
            assert len(val2) == 4, 'Invalid length of rho'
            result_count_1[curr_source][val2[0]*2 + val2[2]] += 1

            if order == 'alternate':
                curr_source = 1 - curr_source
            violations.append(val1_num)
            violations.append(val2_num)
    else:
        if 'Actual rho' in lines[i]:
            line = lines[i].strip()
            seg1 = line[line.find('[')+1:line.find(']')].split(' ')
            val1 = []
            for s in seg1:
                if s != '':
                    val1.append(float(s) < 0)
            assert len(val1) == 4, 'Invalid length of rho'
            result_count_0[curr_source][val1[0]*2 + val1[1]*1] += 1

            seg2 = line[line.find('] [')+3:-1].split(' ')
            val2 = []
            for s in seg2:
                if s != '':
                    val2.append(float(s) < 0)
            assert len(val2) == 4, 'Invalid length of rho'
            result_count_1[curr_source][val2[2]*2 + val2[3]] += 1
            
rows = ['from sampler 0', 'from sampler 1']
cols = ['(r0, r1) = 00', '(r0, r1) = 01', '(r0, r1) = 10', '(r0, r1) = 11']
df1 = pd.DataFrame(result_count_0, columns=cols, index=rows)
print('Falsification result in segment 0:\n', df1, '\n')
cols = ['(r2, r3) = 00', '(r2, r3) = 01', '(r2, r3) = 10', '(r2, r3) = 11']
df2 = pd.DataFrame(result_count_1, columns=cols, index=rows)
print('Falsification result in segment 1:\n', df2)
```
