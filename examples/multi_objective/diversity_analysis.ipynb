```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
directory = 'outputs'
all_files = os.listdir(directory)

trial_name = 'result_multi_hri_demab0'
all_files = [f for f in all_files if f.endswith('.csv') and f.startswith(trial_name+'.')]
#mode = sys.argv[3] # multi / single
mode = 'multi'

fig = plt.figure()
ax = fig.add_subplot()
count = 0
fetch_spawn_y = []
human_spawn_x = []
human_dest_x = []
box_y = []
box_x = []

for file in all_files:
    infile = open(directory+'/'+file, 'r')
    lines = infile.readlines()
    if mode == 'single':
        for i in range(1, len(lines)):
            line = lines[i]
            if float(line.split(',')[-1]) < 0 or float(line.split(',')[-2]) < 0 or float(line.split(',')[-3]) < 0 or float(line2.split(',')[-4]) < 0:
                fetch_spawn_y.append(float(line.split(',')[-5]))
                human_spawn_x.append(float(line.split(',')[-6]))
                human_dest_x.append(float(line.split(',')[-7]))
                box_y.append(float(line.split(',')[-8]))
                box_x.append(float(line.split(',')[-9]))
    else:
        for i in range(1, len(lines), 2):
            line1 = lines[i]
            line2 = lines[i+1]
            if float(line1.split(',')[-1]) < 0 or float(line1.split(',')[-2]) < 0 or float(line2.split(',')[-3]) < 0 or float(line2.split(',')[-4]) < 0:
                fetch_spawn_y.append(float(line1.split(',')[-5]))
                human_spawn_x.append(float(line1.split(',')[-6]))
                human_dest_x.append(float(line1.split(',')[-7]))
                box_y.append(float(line1.split(',')[-8]))
                box_x.append(float(line1.split(',')[-9]))

ax.scatter(fetch_spawn_y, human_dest_x)
ax.set_xlabel('FETCH_SPAWN_Y')
ax.set_ylabel('HUMAN_DEST_X')
#plt.savefig(directory+'/'+sys.argv[2]+'_scatter.png')
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(projection='3d')
ax1.scatter(box_x, box_y, human_spawn_x)
ax1.set_xlabel('box_x')
ax1.set_xlabel('box_y')
ax1.set_xlabel('human_spawn_x')
plt.show()

#print("Variance of ego_speed:", np.var(ego_speed), len(ego_speed))
#print("Variance of ego_brake:", np.var(ego_brake), len(ego_brake))
#print("Variance of adv_speed:", np.var(adv_speed), len(adv_speed))
#print("Variance of adv3_speed:", np.var(adv3_speed), len(adv3_speed))
#print()
```
